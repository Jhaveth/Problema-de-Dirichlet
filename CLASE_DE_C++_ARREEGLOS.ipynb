#Ejemplo 1: Escribe un programa que defina un vector de números y calcule la suma de sus elementos

```python
# Paso 2: Escribir el código C++
%%writefile ejemplo1.cpp
#include <iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    int numeros[5] = {1, 2, 3, 4, 5};
    int suma = 0;
    for (int i = 0; i < 5; i++) {
      suma += numeros[i];
    }
    cout << "La suma de los numeros del vector es: " << suma ;
    return 0;
    //getch(); se deberá usar en c++
}
```

```python
# Paso 3: Compilar el código
%%shell
g++ ejemplo1.cpp -o mi_programa.cpp
```

# Ejercicio 2: Escribe un programa que defina un vector de números y muestre en la salida estándar el vector en orden inverso del último al primer elemento.

# EJEMPLO 3: Desarrolle un programa que lea de la entrada estandar de un vector de enteros y determine el mayor elemento del vector

```python
%%writefile ejemplo3.cpp
#include <iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    int numeros[100];
    cout << "Ingrese el numero de elementos del vector: ";
    int n;
    cin >> n;

    for (int i = 0; i < n; i++) {
      cout<<i+1<< ". Ingrese los elementos del vector: ";
      cin >> numeros[i];//guarda los elementos del vector
    }
    int mayor = numeros[0];
    for (int i = 1; i < n; i++) {
      if (numeros[i] > mayor) {
        mayor = numeros[i];
      }
    }
    cout << "El mayor elemento del vector es: " << mayor << endl;
    return 0;
    //getch(); se deberá usar en c++
}
```

```python
%%shell
g++ ejemplo3.cpp -o mi_programa.cpp
```

#EJEMPLO 4: Realiza un programa que solicite dos vectores de caracteres y después almacene el 2 contenido de ambos vectores en un nuevo vector, situando en primer lugar los elementos del primer vector seguido por los elementos del segundo vector. Muestre el contenido 4 del nuevo vector en la salida estándar

```python
%%writefile ejemplo4.cpp
#include <iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    char vector1[100];
    char vector2[100];
    char vector3[200];
    cout << "Ingrese el numero de elementos del vector 1: ";
    int n1;
    cin >> n1;
    cout << "Ingrese el numero de elementos del vector 2: ";
    int n2;
    cin >> n2;
    cout << "El vector 3 tendra " << n1+n2 << " elementos"<< endl;
    cout<<endl;
    for (int i = 0; i < n1; i++) {
      cout << "Ingrese el caracter No. "<<i+1<< "del vector 1: ";
      cin >> vector1[i];//guarda los elementos del vector
    }
    cout<<endl;
    for (int i = 0; i < n2; i++) {
      cout << "Ingrese el caracter No. "<<i+1<< "del vector2: ";
      cin >> vector2[i];//guarda los elementos del vector
    }
    cout<<endl;
    for (int i = 0; i < n1; i++) {
      vector3[i] = vector1[i];
    }
    for (int i = 0; i < n2; i++) {
      vector3[n1+i] = vector2[i];
    }
    for (int i = 0; i < n1+n2; i++) {
      cout << "El elemento " << i << " del vector 3 es: " << vector3[i] << endl;
    }
    return 0;
    //getch(); se deberá usar en c++
}
```

```python
%%shell
g++ ejemplo4.cpp -o mi_programa.cpp
```
